```python
# Importing Libraries
import rawpy as rp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as patches

from pathlib import Path 
import glob
import os
import sys
```

```python
raw = rp.imread('Stacking_Database/IMG_20240227_113829.dng')
```

```python
bayer = raw.raw_image
bayer_pattern = ''.join(np.array(list((raw.color_desc).decode('ascii')))[raw.raw_pattern].flatten())
color = bayer_pattern.lower()

plt.figure(figsize=(2,2))
gs1 = gridspec.GridSpec(2,2)
gs1.update(wspace=0.05, hspace=0.05)
plt.suptitle('Bayer Pattern', fontsize=14)

for i in range(4):
    ax1 = plt.subplot(gs1[i])
    ax1.set_facecolor(color[i])
    ax1.xaxis.set_visible(False)
    ax1.yaxis.set_visible(False)
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])
    ax1.set_aspect('equal')
    
plt.show()   
```

```python
# The image above tells us the red channel position.  From here, we use the following conditional to extract red and green 
# channels separately = []
red = []

for i in range(bayer.shape[0]):
    if i % 2 != 0:
        red.append(bayer[i][1::2])

green1, green2 = [],[]

for i in range(bayer.shape[0]):
    if i % 2 == 0:
        green1.append(bayer[i][1::2])
    else:
        green2.append(bayer[i][0::2])
```

```python
# Post pixel pushing to create new image capable files after extracting
red = np.concatenate(red, axis=0)
red.shape = (int(bayer.shape[0]/2), int(bayer.shape[1]/2))

green1 = np.concatenate(green1, axis=0)
green1.shape = (int(bayer.shape[0]/2), int(bayer.shape[1]/2))

green2 = np.concatenate(green2, axis=0)
green2.shape = (int(bayer.shape[0]/2), int(bayer.shape[1]/2))

green = (green1+green2)/2
green = green.astype(int)
```

```python
# Selecting the region of interest on the monochrome image
w1, w2 = 400, 400
x1, y1, wx1, wy1 = 790, 550, w1, w2
```

```python
# Plots the monochrome image and highlights the test region
fig, ax = plt.subplots(figsize = (8, 6))
ax.imshow(green,cmap = 'gray')
rect1 = patches.Rectangle((x1, y1), wx1, wy1, linewidth = 1, edgecolor = 'red', facecolor = 'none')
ax.add_patch(rect1)
plt.grid()
plt.show()
```

```python
# %% [markdown]
# ## Helper functions for curvature & banding removal

# %%
def estimate_curvature_field(im):
    """
    Estimate horizontal illumination curvature as a quadratic fit
    to the row-averaged profile, and build a 2D field.
    """
    H, W = im.shape
    x = np.arange(W)
    profile = im.mean(axis=0)
    coeffs = np.polyfit(x, profile, deg=2)
    profile_fit = np.polyval(coeffs, x)
    curv_field = np.tile(profile_fit, (H, 1))
    return curv_field, profile_fit

def estimate_row_banding(im_detrended):
    """
    Estimate rolling-shutter banding pattern from a detrended image
    by looking at row-wise means and removing a low-order vertical trend.
    """
    H, W = im_detrended.shape
    rows = np.arange(H)
    row_means = im_detrended.mean(axis=1)
    coeffs_vert = np.polyfit(rows, row_means, deg=2)
    vert_trend = np.polyval(coeffs_vert, rows)
    row_banding = row_means - vert_trend
    return row_banding

def compute_residual_std(im, curv_field, row_banding):
    """
    Compute the std of the residual after subtracting:
      - curvature field (2D)
      - banding field (2D from row_banding)
    """
    H, W = im.shape
    detrended = im - curv_field
    band_field = np.tile(row_banding[:, None], (1, W))
    residual = detrended - band_field
    res_flat = (residual - residual.mean()).ravel()
    return res_flat.std()
```

```python
im_single = green[y1:y1+w1, x1:x1+w2]

# %% [markdown]
# ## Compute single-frame high-frequency residual std ()

# %%
# Estimate curvature & banding from the single ROI itself
curv_field, profile_fit = estimate_curvature_field(im_single)
im_detrended = im_single - curv_field
row_banding = estimate_row_banding(im_detrended)

sigma_single = compute_residual_std(im_single, curv_field, row_banding)
mean_dn = im_single.mean()

print(f"Single-frame HF residual std (): {sigma_single:.2f} DN")
print(f"Mean DN in ROI: {mean_dn:.2f} DN")
print(f" as % of mean: {100 * sigma_single / mean_dn:.2f} %")
```

```python
# %% [markdown]
# ## Model 1/N decay of camera-induced noise (N = 1..5)

# %%
def model_variable_noise_decay(sigma_single, max_N=5):
    """
    Model the decay of purely variable camera noise under stacking.

    sigma_single : std (DN) from single frame, assumed 100% variable.
    max_N       : maximum number of stacked frames to model.
    """
    N_values = np.arange(1, max_N + 1)
    sigma_model = sigma_single / np.sqrt(N_values)
    return N_values, sigma_model

N_values, sigma_model = model_variable_noise_decay(sigma_single, max_N=5)

print("Modeled variable noise std (assuming all HF residual is camera noise):")
for N, s in zip(N_values, sigma_model):
    print(f"N = {N}:  {s:.2f} DN  ({100*s/mean_dn:.2f} % of mean)")

plt.figure(figsize=(7, 5))
plt.plot(N_values, sigma_model, 'o-', label='Modeled (N) =  / N')
plt.xlabel('Number of stacked frames (N)')
plt.ylabel('Modeled HF noise std [DN]')
plt.title('Modeled reduction of high-frequency noise\n(assuming all HF residual is camera noise)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Load file list
dng_paths = glob.glob(r"Stacking_Database/*.dng")
dng_paths.sort()

frames = []

for path in dng_paths:
    with rp.imread(path) as raw:
        # Get the Bayer mosaic (linear sensor-level RAW)
        bayer = raw.raw_image_visible.astype(np.float32)

    # --- Extract both green channels from the Bayer mosaic ---
    g1_rows = []
    g2_rows = []

    for i in range(bayer.shape[0]):
        if i % 2 == 0:
            # Even rows  green on odd columns
            g1_rows.append(bayer[i][1::2])
        else:
            # Odd rows  green on even columns
            g2_rows.append(bayer[i][0::2])

    # Convert lists to arrays
    g1 = np.array(g1_rows, dtype=np.float32)
    g2 = np.array(g2_rows, dtype=np.float32)

    # --- Average the two green planes ---
    G = (g1 + g2) / 2.0   # THIS is the new key line

    # --- Extract ROI ---
    im_roi = G[y1:y1+w1, x1:x1+w2]
    frames.append(im_roi)

# Stack into a single array: shape (N_frames, H, W)
frames = np.stack(frames, axis=0)

print("frames shape:", frames.shape)
```

```python
Nframes, H, W = frames.shape

# Full stack (average of all frames)
im_stack_full = frames.mean(axis=0)
mean_dn = im_stack_full.mean()

# Curvature from full stack
curv_field, profile_fit = estimate_curvature_field(im_stack_full)

# Banding from full stack (after curvature removal)
stack_detrended = im_stack_full - curv_field
row_banding = estimate_row_banding(stack_detrended)

print("Mean DN (full stack ROI):", mean_dn)
```

```python
def compute_residual_std(im, curv_field, row_banding):
    """Std of residual after subtracting curvature and banding."""
    H, W = im.shape
    detrended = im - curv_field
    band_field = np.tile(row_banding[:, None], (1, W))
    residual = detrended - band_field
    res_flat = (residual - residual.mean()).ravel()
    return res_flat.std()

Ns_actual = np.arange(1, Nframes + 1)
sigma_actual = []

for N in Ns_actual:
    im_stack_N = frames[:N].mean(axis=0)
    sN = compute_residual_std(im_stack_N, curv_field, row_banding)
    sigma_actual.append(sN)

sigma_actual = np.array(sigma_actual)

print("Actual residual (N) from real stacks:")
for N, s in zip(Ns_actual, sigma_actual):
    print(f"N = {N}:  {s:.2f} DN  ({100*s/mean_dn:.2f} % of mean)")
```

```python
# Contrast a purely stochastic model with actual data
sigma_single_measured = sigma_actual[0]  #  at N=1 from real data
N_values_model, sigma_model = model_variable_noise_decay(sigma_single_measured, max_N=Nframes)

plt.figure(figsize=(7, 5))

# Actual measurements
plt.plot(Ns_actual, sigma_actual, 'o-', label='Actual (N) (real stacks)')

# Modeled camera-only noise decay
plt.plot(N_values_model, sigma_model, 's--', label='Modeled (N) =  / N (camera-only)')

plt.xlabel('Number of stacked frames (N)')
plt.ylabel('Residual std (N) [DN]')
plt.title('Residual width vs. number of stacked frames')
plt.grid(True)
plt.legend()
plt.show()
```
